```python
def emissor_coletor(hoe,hfe,hre,hie):
  hic = hie
  hrc = 1 - hre
  hfc = -(1+hfe)
  hoc = hoe
  return hic,hrc,hfc,hoc
```

```python
def emissor_base(hoe,hfe,hre,hie):
  hib = hie/((1+hfe)*(1-hre) + hie*hoe)
  hrb = (hie*hoe - hre*(1+hfe))/((1+hfe)*(1-hre) + hie*hoe)
  hfb = (-hfe*(1-hre) - hie*hoe)/((1+hfe)*(1-hre) + hie*hoe)
  hob = hoe/((1+hfe)*(1-hre) + hie*hoe)
  return hib,hrb,hfb,hob
```
